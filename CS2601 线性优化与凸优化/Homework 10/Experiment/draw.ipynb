```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 100)
y1 = -1 * x + 3
y2 = 2 * x - 1
y3 = -8 / 3 * x + 47 / 9

figure = plt.figure()
plt.plot(x, y1, label=r'$\mu_2=-\mu_1+3$')
plt.plot(x, y2, label=r'$\mu_2=2\mu_1-1$')
plt.plot(x, y3, label=r'$-8\mu_1-3\mu_2=-\frac{47}{3}$')
plt.scatter(4 / 3, 5 / 3, c='red', s=100, marker='*', zorder=5)
plt.xlim((0, 4))
plt.ylim((0, 6))
plt.xlabel(r'$\mu_1$')
plt.ylabel(r'$\mu_2$')
y4 = np.minimum(y1, y2)
plt.fill_between(x, y4, y2, where=y4<y2, color='grey', alpha=0.3)
plt.grid(True)
plt.legend(loc=1)
figure.savefig('./figures/p2.png')
plt.show()
```
